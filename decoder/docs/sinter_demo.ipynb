```python
def H(n, if_self_dual=False):
    d = length_dist_dict[n]
    F2 = galois.GF(2)
    alistFilePath = alistDirPath + "n" + str(n) + "_d" + str(d) + ".alist"
    GenMat = F2(readAlist(alistFilePath))
    
    if if_self_dual:
        G_punctured = GenMat[:, :-1]  # Puncture the last column
        # Parity-check matrix of the punctured code
        H_punctured = G_punctured.null_space() 
    else:
        # For GenMat being a dual-containing code, use null-space as parity-check
        H_punctured = GenMat.null_space() 

    # --- Construct Hx and Hz ---
    # For this construction, we use the same matrix for both X and Z checks
    Hx = H_punctured
    Hz = H_punctured

    # Convert from Galois Field array to standard Numpy uint8 array
    # This is crucial for passing them into Stim/Sinter functions
    Hx = np.array(Hx, dtype=np.uint8)
    Hz = np.array(Hz, dtype=np.uint8)

    return Hx, Hz
```

```python
def find_logical_z_operator(Hx, Hz):
    """
    Finds a Logical Z operator (L_Z) for a CSS code.
    L_Z must commute with Hx and be independent of Hz.
    """
    num_qubits = Hx.shape[1]
    
    # Simple randomized search for valid L_Z
    # 1. Transversal Z (all 1s) is a common candidate for color codes / some QR codes
    candidates = [np.ones(num_qubits, dtype=np.uint8)]
    
    # 2. Add random candidates
    for _ in range(100):
        candidates.append(np.random.randint(0, 2, size=num_qubits).astype(np.uint8))

    for candidate in candidates:
        # Check commutation with X stabilizers (Hx @ candidate % 2 == 0)
        if np.any((Hx @ candidate) % 2):
            continue
            
        # Check independence from Z stabilizers (Hz) using simple rank check
        # If Rank([Hz]) < Rank([Hz, candidate]), then candidate is independent
        m_stack = np.vstack([Hz, candidate])
        
        # Simple GF(2) rank calculation
        def gf2_rank(rows):
            rows = np.array(rows, copy=True)
            pivot_row = 0
            num_rows, num_cols = rows.shape
            for col in range(num_cols):
                if pivot_row >= num_rows: break
                if rows[pivot_row, col] == 0:
                    swap = np.argmax(rows[pivot_row:, col]) + pivot_row
                    if rows[swap, col] == 0: continue
                    rows[[pivot_row, swap]] = rows[[swap, pivot_row]]
                for r in range(pivot_row + 1, num_rows):
                    if rows[r, col]: rows[r] ^= rows[pivot_row]
                pivot_row += 1
            return pivot_row

        if gf2_rank(m_stack) > gf2_rank(Hz):
            return candidate

    # Fallback: Use Stim to find logicals (more robust)
    # Construct a Tableau representing the code state and ask for logicals
    # This is complex to implement from scratch, so if the random search fails,
    # the code likely has no logical qubits or is defined incorrectly.
    print("WARNING: Could not find logical Z. Using placeholder (Observable 0 will be random).")
    return np.zeros(num_qubits, dtype=np.uint8)
```

```python
import stim

def generate_css_memory_experiment(
    Hx: np.ndarray, 
    Hz: np.ndarray, 
    rounds: int,
) -> stim.Circuit:
    check_css_validity(Hx, Hz)

    num_data = Hx.shape[1]
    num_x_checks = Hx.shape[0]
    num_z_checks = Hz.shape[0]
    
    data_qubits = list(range(num_data))
    x_ancillas = list(range(num_data, num_data + num_x_checks))
    z_ancillas = list(range(num_data + num_x_checks, num_data + num_x_checks + num_z_checks))
    
    circuit = stim.Circuit()
    circuit.append("R", data_qubits + x_ancillas + z_ancillas)
    circuit.append("TICK")
    
    def append_measurement_round(c: stim.Circuit, is_first_round: bool):
        c.append("H", x_ancillas)
        c.append("TICK")
        for check_idx, row in enumerate(Hx):
            targets = [data_qubits[q] for q in np.flatnonzero(row)]
            ancilla = x_ancillas[check_idx]
            for t in targets:
                c.append("CX", [ancilla, t])
        c.append("TICK")
        c.append("H", x_ancillas)
        c.append("TICK")
        c.append("M", x_ancillas)
        c.append("TICK")
        
        for check_idx, row in enumerate(Hz):
            targets = [data_qubits[q] for q in np.flatnonzero(row)]
            ancilla = z_ancillas[check_idx]
            for t in targets:
                c.append("CX", [t, ancilla])
        c.append("TICK")
        c.append("M", z_ancillas)
        c.append("TICK")
        c.append("R", x_ancillas + z_ancillas)
        c.append("TICK")

        total_measurements_per_round = num_x_checks + num_z_checks
        
        # X-Checks
        for i in range(num_x_checks):
            current_rec = stim.target_rec(-total_measurements_per_round + i)
            if not is_first_round:
                prev_rec = stim.target_rec(-total_measurements_per_round * 2 + i)
                c.append("DETECTOR", [current_rec, prev_rec], [i, 0, 0])
                c.append("TICK")
            # Note: No detector for first round X-check (gauge)

        # Z-Checks
        for i in range(num_z_checks):
            offset_in_round = num_x_checks + i
            current_rec = stim.target_rec(-total_measurements_per_round + offset_in_round)
            if not is_first_round:
                prev_rec = stim.target_rec(-total_measurements_per_round * 2 + offset_in_round)
                c.append("DETECTOR", [current_rec, prev_rec], [num_x_checks + i, 0, 0])
                c.append("TICK")
            else:
                c.append("DETECTOR", [current_rec], [num_x_checks + i, 0, 0])
                c.append("TICK")    

    loop_body = stim.Circuit()
    append_measurement_round(loop_body, is_first_round=False)
    append_measurement_round(circuit, is_first_round=True)
    
    if rounds > 1:
        circuit.append(stim.CircuitRepeatBlock(rounds - 1, loop_body))

    # --- 3. Final Readout ---
    circuit.append("M", data_qubits)
    circuit.append("TICK")
    
    for i, row in enumerate(Hz):
        rec_targets = []
        data_indices = np.flatnonzero(row)
        for data_idx in data_indices:
            offset = -(num_data - data_idx)
            rec_targets.append(stim.target_rec(offset))
        z_ancilla_offset = -(num_data + num_z_checks - i)
        rec_targets.append(stim.target_rec(z_ancilla_offset))
        circuit.append("DETECTOR", rec_targets, [num_x_checks + i, 1, 0])
        circuit.append("TICK")

    # # --- 4. Append Logical Observable ---
    # L_Z_bits = find_logical_z_operator(Hx, Hz)
    
    # obs_targets = []
    # # Data qubits are the last 'num_data' measurements
    # for k in np.flatnonzero(L_Z_bits):
    #     offset = -(num_data - k)
    #     obs_targets.append(stim.target_rec(offset))
        
    # circuit.append("OBSERVABLE_INCLUDE", obs_targets, 0)
    # circuit.append("TICK")
    
    return circuit
```

```python
import stim
import numpy as np

def standard_depolarizing_noise_model(
        circuit: stim.Circuit,
        data_qubits: list[int],
        after_clifford_depolarization: float,
        after_reset_flip_probability: float,
        before_measure_flip_probability: float,
        before_round_data_depolarization: float
) -> stim.Circuit:
    result = stim.Circuit()
    first_reset_seen = False
    tick_count = 0

    for instruction in circuit:
        if isinstance(instruction, stim.CircuitRepeatBlock):
            result.append(stim.CircuitRepeatBlock(
                repeat_count=instruction.repeat_count,
                body=standard_depolarizing_noise_model(
                    instruction.body_copy(),
                    data_qubits,
                    after_clifford_depolarization,
                    after_reset_flip_probability,
                    before_measure_flip_probability,
                    before_round_data_depolarization
                )))
        elif instruction.name == 'R' and not first_reset_seen:
            result.append(instruction)
            result.append('X_ERROR', instruction.targets_copy(), after_reset_flip_probability)
            first_reset_seen = True
        elif instruction.name in ['CNOT', 'CX', 'CZ']:
            result.append(instruction)
            result.append('DEPOLARIZE2', instruction.targets_copy(), after_clifford_depolarization)
        elif instruction.name == 'MR':
            result.append('X_ERROR', instruction.targets_copy(), before_measure_flip_probability)
            result.append(instruction)
            result.append('X_ERROR', instruction.targets_copy(), after_reset_flip_probability)
        elif instruction.name == 'M':
            result.append('X_ERROR', instruction.targets_copy(), before_measure_flip_probability)
            result.append(instruction)
        elif instruction.name == 'TICK':
            result.append(instruction)
            tick_count += 1
            # Assuming a standard surface code schedule where a round is ~9 ticks? 
            # Note: Your specific circuit might not follow the 9-tick structure exactly.
            # This logic adds idle noise at start of rounds.
            if first_reset_seen and before_round_data_depolarization > 0:
                 # Simple heuristic: Apply if it looks like the start of a round block
                 # or simply apply it if your schedule relies on TICKs.
                 # Original code logic: if tick_count >= 2 and (tick_count - 1) % 9 == 1:
                 # Adjusting to apply generally for this example:
                 pass 
        else:
            result.append(instruction)
            
    return result

def si1000_noise_model(
        circuit: stim.Circuit,
        data_qubits: list[int],
        probability: float
) -> stim.Circuit:
    all_qubits_in_circuit = set()
    # Pre-scan to find all used qubits if not provided
    for op in circuit.flattened():
        for t in op.targets_copy():
            if t.value is not None: all_qubits_in_circuit.add(t.value)
    
    result = stim.Circuit()
    first_reset_seen = False
    
    for instruction in circuit:
        if isinstance(instruction, stim.CircuitRepeatBlock):
            result.append(stim.CircuitRepeatBlock(
                repeat_count=instruction.repeat_count,
                body=si1000_noise_model(instruction.body_copy(), data_qubits, probability)
            ))
        elif instruction.name == 'R' and not first_reset_seen:
            result.append(instruction)
            result.append('X_ERROR', instruction.targets_copy(), 2 * probability)
            first_reset_seen = True
        elif instruction.name in ['CNOT', 'CX']:
            result.append(instruction)
            result.append('DEPOLARIZE2', instruction.targets_copy(), probability / 10)
        elif instruction.name == 'MR':
            # Split MR into noise components
            result.append('DEPOLARIZE1', instruction.targets_copy(), probability) # Before measurement
            result.append('X_ERROR', instruction.targets_copy(), 5 * probability) # Meas flip
            result.append(instruction)
            result.append('X_ERROR', instruction.targets_copy(), 2 * probability) # Reset error
        elif instruction.name == 'M':
            result.append('DEPOLARIZE1', instruction.targets_copy(), probability)
            result.append('X_ERROR', instruction.targets_copy(), 5 * probability)
            result.append(instruction)
        else:
            result.append(instruction)
            
    return result

def bravyi_noise_model(
        circuit: stim.Circuit,
        error_rate: float
) -> stim.Circuit:
    result = stim.Circuit()
    first_reset_seen = False
    
    for instruction in circuit:
        if isinstance(instruction, stim.CircuitRepeatBlock):
            result.append(stim.CircuitRepeatBlock(
                repeat_count=instruction.repeat_count,
                body=bravyi_noise_model(instruction.body_copy(), error_rate)
            ))
        elif instruction.name == 'R' and not first_reset_seen:
            result.append(instruction)
            result.append('X_ERROR', instruction.targets_copy(), error_rate)
            first_reset_seen = True
        elif instruction.name == 'RX':
            result.append(instruction)
            result.append('Z_ERROR', instruction.targets_copy(), error_rate)
        elif instruction.name == 'RZ':
            result.append(instruction)
            result.append('X_ERROR', instruction.targets_copy(), error_rate)
        elif instruction.name == 'MX':
            result.append('Z_ERROR', instruction.targets_copy(), error_rate)
            result.append(instruction)
        elif instruction.name == 'MZ':
            result.append('X_ERROR', instruction.targets_copy(), error_rate)
            result.append(instruction)
        elif instruction.name == 'I':
            result.append(instruction)
            result.append('DEPOLARIZE1', instruction.targets_copy(), error_rate)
        elif instruction.name in ['CNOT', 'CX', 'CZ']:
            result.append(instruction)
            result.append('DEPOLARIZE2', instruction.targets_copy(), error_rate)
        elif instruction.name == 'M':
            result.append('X_ERROR', instruction.targets_copy(), error_rate)
            result.append(instruction)
        else:
            result.append(instruction)
    return result
```

```python
def generate_experiment_with_noise(
    Hx: np.ndarray, 
    Hz: np.ndarray, 
    rounds: int, 
    noise_model_name: str,
    noise_params: dict = None
) -> stim.Circuit:
    """
    Generates a circuit and applies the specified noise model.
    If noise_params is None, defaults to p=0.001 for all operations.
    """
    # 1. Generate clean circuit (noise_p=0)
    clean_circuit = generate_css_memory_experiment(Hx, Hz, rounds, noise_p=0.0)
    
    # 2. Identify data qubits
    num_data = Hx.shape[1]
    data_qubits = list(range(num_data))
    
    # --- HELPER: robust parameter extraction ---
    def get_p(params, default=0.001):
        if params is None: return default
        # Look for common keys
        for key in ['p', 'p_clifford', 'p_gate', 'noise', 'error_rate']:
            if key in params: return params[key]
        return default

    # 3. Apply Noise Model
    # We extract a "base_p" to use as a fallback for any missing specific parameter.
    base_p = get_p(noise_params)
    
    # If params is None, treat it as an empty dict for .get() calls
    params = noise_params if noise_params is not None else {}

    if noise_model_name == "depolarizing":
        return standard_depolarizing_noise_model(
            circuit=clean_circuit,
            data_qubits=data_qubits,
            after_clifford_depolarization=params.get('p_clifford', base_p),
            after_reset_flip_probability=params.get('p_reset', base_p),
            before_measure_flip_probability=params.get('p_meas', base_p),
            before_round_data_depolarization=params.get('p_data_round', base_p)
        )
        
    elif noise_model_name == "si1000":
        # SI1000 derives everything from a single p
        return si1000_noise_model(
            circuit=clean_circuit,
            data_qubits=data_qubits,
            probability=base_p
        )
        
    elif noise_model_name == "bravyi":
        # Bravyi uses a single error rate for everything
        return bravyi_noise_model(
            circuit=clean_circuit,
            error_rate=base_p
        )
    else:
        raise ValueError(f"Unknown noise model: {noise_model_name}")
```

```python
import sinter
import os
from mwpf.sinter_decoders import SinterMWPFDecoder
from typing import List

length_dist_dict = {7:3, 23:7, 47:11} 

noise_values = [0.008, 0.009, 0.01, 0.011, 0.012]

# Create Sinter Tasks
tasks = []
for n in [7]:
    d = length_dist_dict[n]
    Hx, Hz = H(n)
    
    for p in noise_values:
        tasks.append(
            sinter.Task(
                circuit = generate_experiment_with_noise(
                    Hx=Hx,
                    Hz=Hz,
                    rounds=d * 3,
                    noise_model_name="si1000",
                    noise_params={"p": p}
                ),
                decoder='mwpf',
                json_metadata={'n': n, 'd': d, 'r': d * 3, 'p': p},
            )
        )

collected_stats: List[sinter.TaskStats] = sinter.collect(
    num_workers=os.cpu_count(),
    tasks=tasks,
    custom_decoders = { "mwpf": SinterMWPFDecoder(cluster_node_limit=15) },
    max_shots=100_000,
    print_progress=True,
)
```

```python
import sinter
import tempfile
import os
import pandas as pd
from typing import List

# Helper to parse the binary data into the format your plot code expects
def parse_and_average_stats(stats: List[sinter.TaskStats], trace_file: str):
    # Use the static method already in your provided class
    df_details = SinterMWPFDecoder.parse_mwpf_trace(trace_file)
    
    # Map the detailed shot-by-shot stats back to the sinter tasks
    # (Simplified: assuming one trace file contains stats for all tasks in order)
    # For a more robust approach, create one trace file per sinter task.
    
    results = []
    # Note: If shots are interleaved, you'll need to index by task metadata
    # Here we calculate global means for demonstration
    avg_cpu = df_details['cpu_time'].mean()
    avg_obj = df_details['objective_value'].mean()
    
    for s in stats:
        m = s.json_metadata
        results.append({
            'n': m['n'], 'p': m['p'], 'shots': s.shots,
            'total_logical_error_rate': s.errors / s.shots,
            'mean_objective_value': avg_obj,
            'average_cpu_time_seconds': avg_cpu,
            # Add placeholders for CI columns expected by your plot script
            'total_err_low': 0, 'total_err_high': 0,
            'x_logical_error_rate': 0, 'z_logical_error_rate': 0
        })
    return pd.DataFrame(results)

# --- Execution ---

# 1. Create a temporary file for the trace
with tempfile.NamedTemporaryFile(delete=False, suffix=".bin") as tmp:
    trace_path = tmp.name

# 2. Configure the decoder with the trace path
mwpf_decoder = SinterMWPFDecoder(
    cluster_node_limit=15,
    trace_filename=trace_path
)

# 3. Run Sinter
collected_stats = sinter.collect(
    num_workers=os.cpu_count(),
    tasks=tasks,
    custom_decoders={"mwpf": mwpf_decoder},
    max_shots=100_000,
    print_progress=True,
)

# 4. Convert stats to your Plot-compatible CSV
final_df = parse_and_average_stats(collected_stats, trace_path)
final_df.to_csv("hyperion_results.csv", index=False)

# Clean up
if os.path.exists(trace_path):
    os.remove(trace_path)
```

```python
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats

## CONFIGURATION: Set your parameters here
CSV_FILEPATH = "hyperion_results.csv"  # Path to your generated CSV
ALPHA = 0.05                           # 95% Confidence Interval
length_dist_dict = {7:3, 17:5, 23:7, 47:11, 79:15, 103:19, 167:23}

## DATA LOADING & PRE-PROCESSING
if not os.path.exists(CSV_FILEPATH):
    print(f"Error: File '{CSV_FILEPATH}' not found.")
else:
    df = pd.read_csv(CSV_FILEPATH)
    
    # Attempt to extract metadata from filename if missing in CSV
    bias_match = re.search(r"bias([\d\.]+)", CSV_FILEPATH)
    bias_factor = float(bias_match.group(1)) if bias_match else "Unknown"
    
    shots_match = re.search(r"shots(\d+)", CSV_FILEPATH)
    shots_val = int(shots_match.group(1)) if shots_match else (df['shots'].iloc[0] if 'shots' in df.columns else 100000)

    ns = sorted(df['n'].unique())
    ps = sorted(df['p'].unique())
    confidence_percent = int((1 - ALPHA) * 100)

    # PLOT STYLING & INITIALIZATION
    plt.rcParams.update({
        'font.size': 14, 'axes.titlesize': 18, 'axes.labelsize': 16,     
        'xtick.labelsize': 14, 'ytick.labelsize': 14, 'legend.fontsize': 12
    })

    fig = plt.figure(figsize=(18, 12))
    gs = fig.add_gridspec(2, 6, height_ratios=[3, 2])
    
    # Modified layout: Total error takes the whole top row
    ax_total = fig.add_subplot(gs[0, :])
    ax_obj = fig.add_subplot(gs[1, 0:3])
    ax_time = fig.add_subplot(gs[1, 3:6])
    
    ler_axes = [ax_total]
    plot_configs = [
        {'key': 'total_logical_error_rate', 'title': 'Total Logical Error'}
    ]

    # PLOT LOGICAL ERROR RATES
    for i, config in enumerate(plot_configs):
        ax = ler_axes[i]
        key = config['key']
        
        for n in ns:
            subset = df[df['n'] == n].sort_values('p')
            if subset.empty or key not in subset.columns: continue
            
            d = length_dist_dict.get(n, '?')
            p_L = subset[key].values
            
            # Inline Confidence Interval Calculation (Clopper-Pearson)
            k = p_L * shots_val
            lower_b = scipy.stats.beta.ppf(ALPHA / 2, k, shots_val - k + 1)
            lower_b[k == 0] = 0.0
            upper_b = scipy.stats.beta.ppf(1 - ALPHA / 2, k + 1, shots_val - k)
            upper_b[k == shots_val] = 1.0
            
            y_err = [p_L - lower_b, upper_b - p_L]

            ax.errorbar(subset['p'], p_L, yerr=y_err, marker='o', linestyle='-', 
                        label=f'n={n}, d={d}', capsize=3, alpha=0.8)

        # Break-even and Resolution Limit
        ax.plot(ps, ps, linestyle='--', color='black', alpha=0.4, label="Break-even")
        res_limit = 1.0 / shots_val
        ax.axhline(y=res_limit, color='gray', linestyle=':', alpha=0.5)
        
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.set_xlabel('Physical Error Rate (p)')
        ax.set_title(config['title'])
        ax.grid(True, which="both", ls="--", alpha=0.3)
        if i == 0: ax.set_ylabel(f'LER ({confidence_percent}% CI)')

## PLOT MEAN OBJECTIVE & CPU TIME
    # Objective Plot
    if 'mean_objective_value' in df.columns:
        for n in ns:
            subset = df[df['n'] == n].sort_values('p')
            ax_obj.plot(subset['p'], subset['mean_objective_value'], marker='s', linestyle=':', label=f'n={n}')
        ax_obj.set_xscale('log')
        ax_obj.set_xlabel('Physical Error Rate (p)')
        ax_obj.set_ylabel('Weight')
        ax_obj.set_title('Mean Decoder Objective Value')
        ax_obj.grid(True, which="both", ls="--", alpha=0.3)

    # CPU Time Plot
    if 'average_cpu_time_seconds' in df.columns:
        for n in ns:
            subset = df[df['n'] == n].sort_values('p')
            ax_time.plot(subset['p'], subset['average_cpu_time_seconds'], marker='^', linestyle='-.')
        ax_time.set_xscale('log')
        ax_time.set_yscale('log')
        ax_time.set_xlabel('Physical Error Rate (p)')
        ax_time.set_ylabel('Seconds / Shot')
        ax_time.set_title('Avg Computation Time')
        ax_time.grid(True, which="both", ls="--", alpha=0.3)

    # Final Layout Touches
    handles, labels = ax_total.get_legend_handles_labels()
    fig.legend(handles, labels, loc='center right', bbox_to_anchor=(1.05, 0.5), title="Code Parameters")
    fig.suptitle(f'Hyperion Decoder Performance (Bias: {bias_factor}, Shots: {shots_val})', fontsize=22)
    
    plt.tight_layout(rect=[0, 0, 0.92, 0.95])
    plt.show()
```
